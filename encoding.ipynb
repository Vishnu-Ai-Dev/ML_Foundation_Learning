```python
#ml runs on numbers so when specific characters comes in we need to encode it as numbers

import pandas as pd
from sklearn.preprocessing import LabelEncoder


#datas

data={"Gender":["Male","Female","Others","Male","Female","Others","Male"]}
df=pd.DataFrame(data)

#encoding the datas we are using we are using label encoding we have particular set of order to be followed

encoder=LabelEncoder()
df["Gender"]=encoder.fit_transform(df["Gender"])

print(df)







```
