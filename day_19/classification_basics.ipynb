The Core Difference: Regression vs. Classification
Instead of predicting continuous numbers (like pricing or salary), we are now shifting to predicting categories and classes.

Linear Regression: Predicts continuous numbers. (e.g., "What will the exact temperature be?")

Logistic Regression: Predicts discrete categories. (e.g., "Will it rain? Yes or No.")

Classification is the backbone of modern AI systems, actively used in:

Spam Detection algorithms

Medical AI (Disease vs. No Disease)

Image Recognition

Types of Classification
1. Binary Classification
Deals with target variables that have exactly two possible outcomes.

True or False

Pass or Fail

Spam or Inbox

2. Multi-Class Classification
Deals with target variables that have more than two possible outcomes.

Predicting an animal type from an image: (Dog, Cat, Horse)

Predicting weather categories: (Sunny, Rainy, Cloudy, Snowy)

```python
import numpy as np

z=3

sigmoid=1/(1+np.exp(-z))    
print(sigmoid)  
```

Let us understand the Machine Learning internally.

The core math behind Logistic Regression actually starts with the exact same linear equation as Linear Regression

The Problem: The raw output of a linear equation ($z$) can be infinitely large (like 20, 100, or -500). If we are trying to predict a probability (which must be between 0 and 1), a massive number breaks the logic.The Solution: We take that raw $z$ value and pass it through the Sigmoid Function to squeeze it.

σ(x) = 1 / (1 + e^(-x))

What does this formula actually do?It takes any mathematical number (from negative infinity to positive infinity) and forces it into a strict range between 0 and 1.If you plot this mathematically, it creates a distinct S-shaped curve.Example in Python: If our linear equation outputs $z = 3$, passing it through the Sigmoid function outputs roughly $0.95$. We can confidently classify this as a "Yes" or "True".
