```python
#understanding of weaker dataset 

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#Dataset with weaker correlation 

X=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([20,90,10,50,40])

#training the model

model=LinearRegression()
model.fit(X,y)

#evaluationg the bad model

pred = model.predict(X)

print("R2 Score:", r2_score(y, pred))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score

#Dataset with better correlation

X=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([10,20,30,40,50])

#training the model

model=LinearRegression()
model.fit(X,y)

#Evaluating the model

pred = model.predict(X)
print("R2 Score:", r2_score(y, pred))
```
